```python
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv('clean_cafe_sales.csv')

# Configurando o estilo dos gráficos 
sns.set_theme(style="darkgrid")
```

### 1. Padrões de Vendas
Pergunta:
"Qual é o item mais vendido em termos de quantidade e receita? Existe sazonalidade nas vendas?"

```python
# Item mais vendido por quantidade 
top_itens_quantidade = df.groupby('Item')['Quantity'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x= top_itens_quantidade, y= top_itens_quantidade.index, palette= 'Blues_d')
plt.title('Item mais vendido por quantidade')
plt.xlabel('Receita total (USD)')
plt.ylabel('Item')
plt.show()
```

#### - Podemos observar que em questão de quantidade de vendas há similaridade entre todos os produtos


```python
# Item com maior receita 
top_itens_receita = df.groupby('Item')['Total Spent'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x= top_itens_receita, y= top_itens_receita.index, palette= 'Blues_d')
plt.title('Item com maior receita')
plt.xlabel('Receita Total (USD)')
plt.ylabel('Item')
plt.show()
```

#### - Agora quando oberservamos a receita observamos resultados discrepantes, Observando que 3 Itens se Destacam liderando a receita total: Salad, Sandwich, Smoothie.

### 2. Comportamento de Pagamento
Pergunta:
"Qual método de pagamento é mais popular?"



```python
# Popularidade por meio de pagamento
contagem_pagamento = df['Payment Method'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(contagem_pagamento, labels=contagem_pagamento.index,autopct='%1.1f%%',
        colors=sns.color_palette('pastel'))
plt.title('Distribuição de Métodos de Pagamento')
plt.show()
```

#### - Podemos notar que A maior parte das vendas ocorre pelo método de pagamento Digital Wallet.

### 3. Preços e Margens
Pergunta:
"Qual item tem a maior contribuição para a receita total?"

```python
# Contribuição percentual de cada item para a receita total
contribuicao_receita = (top_itens_receita/top_itens_receita.sum()) * 188
plt.figure(figsize=(10,6))
sns.barplot(x=contribuicao_receita.values, y=contribuicao_receita.index,
            palette='viridis')
plt.title('Contribuição Percentual por Item a Receita Total')
plt.xlabel('Contribuição(%)')
plt.ylabel('Item')
plt.show()
```

#### - Em contribuição total da receita podemos obervar que Somente Salad representa A maior parte da Receita total.

### 4. Impacto da Limpeza

"Quantos registros foram removidos ou corrigidos durante a limpeza?"
